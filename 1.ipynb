```python
#2ndQuestion

# prompt: uniform cost searc

import heapq

def uniform_cost_search(graph, start, goal):

  priority_queue = [(0, start, [])]  # (cost, node, path)
  visited = set()

  while priority_queue:
    cost, current_node, path = heapq.heappop(priority_queue)

    if current_node == goal:
      return path + [current_node], cost

    if current_node not in visited:
      visited.add(current_node)

      for neighbor, edge_cost in graph.get(current_node, []):
        if neighbor not in visited:
          heapq.heappush(priority_queue, (cost + edge_cost, neighbor, path + [current_node]))

  return None, float('inf')  # No path found


# Example graph (replace with your own)
graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('D', 5), ('E', 12)],
    'C': [('F', 2)],
    'D': [('G', 3)],
    'F': [('G', 1)]
}

start_node = 'A'
goal_node = 'G'

path, cost = uniform_cost_search(graph, start_node, goal_node)

if path:
  print(f"Shortest path from {start_node} to {goal_node}: {path}")
  print(f"Total cost: {cost}")
else:
  print(f"No path found from {start_node} to {goal_node}")




def dfs_iterative(graph, start_node):
    """
    Performs an iterative depth-first search (DFS) on a graph.

    Args:
        graph: A dictionary where keys are nodes and values are lists of adjacent nodes.
        start_node: The node from which DFS traversal starts.

    Returns:
        A list representing the DFS traversal order.
    """
    visited = set()
    stack = [start_node]
    traversal_order = []

    while stack:
        node = stack.pop()

        if node not in visited:
            visited.add(node)
            traversal_order.append(node)

            # Add neighbors in reverse order to maintain DFS order
            neighbors = graph.get(node, [])
            stack.extend(reversed(neighbors))  # Extend is better than looping

    return traversal_order


# Example graph (Undirected)
graph = {
    0: [1, 2],
    1: [0, 3, 5, 2],
    2: [0, 1, 4, 5, 3],
    3: [1, 2],
    4: [2, 5],
    5: [1, 2, 4]
}

start_node = 0
traversal_order = dfs_iterative(graph, start_node)
print(f"DFS traversal order starting from node {start_node}: {traversal_order}")







def bidirectional_search(graph, start_node, goal_node):

    if start_node == goal_node:
        return [start_node]

    queue_forward = [(start_node, [start_node])]
    queue_backward = [(goal_node, [goal_node])]

    visited_forward = {start_node: [start_node]}
    visited_backward = {goal_node: [goal_node]}

    while queue_forward and queue_backward:
        # Expand forward search
        current_node_forward, path_forward = queue_forward.pop(0)
        for neighbor in graph.get(current_node_forward, []):
            if neighbor not in visited_forward:
                new_path_forward = path_forward + [neighbor]
                visited_forward[neighbor] = new_path_forward
                queue_forward.append((neighbor, new_path_forward))

                if neighbor in visited_backward:
                    # Connect the two paths
                    return new_path_forward + visited_backward[neighbor][::-1][1:]

        # Expand backward search
        current_node_backward, path_backward = queue_backward.pop(0)
        for neighbor in graph.get(current_node_backward, []):
            if neighbor not in visited_backward:
                new_path_backward = [neighbor] + path_backward
                visited_backward[neighbor] = new_path_backward
                queue_backward.append((neighbor, new_path_backward))

                if neighbor in visited_forward:
                    # Connect the two paths
                    return visited_forward[neighbor] + new_path_backward[::-1][1:]

    return None  # No path found


# Example bidirectional graph
graph = {
    0: [1, 2],
    1: [0, 3, 5, 2],
    2: [0, 1, 4, 5, 3],
    3: [1, 2],
    4: [2, 5],
    5: [1, 2, 4]
}

start_node = 0
goal_node = 3

path = bidirectional_search(graph, start_node, goal_node)

if path:
    print(f"Shortest path from {start_node} to {goal_node}: {path}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```

```python
#3rd question water jug

def bfs_water_jug(x, y, target,searchAlg):
    visited = set()
    queue = [((0, 0), [])] # Store state and path
    
    while queue:
        (a, b), path = queue.pop(searchAlg)
        # If we reached the target, return the path
        if a == target or b == target or a + b == target:
            return path + [(a, b)]
        
        # Skip if already visited
        if (a, b) in visited:
            continue
        visited.add((a, b))
        
        # Generate all possible next states
        next_states = [
            ((x, b), "Fill Jug 1"),       # Fill jug1
            ((a, y), "Fill Jug 2"),       # Fill jug2
            ((0, b), "Empty Jug 1"),      # Empty jug1
            ((a, 0), "Empty Jug 2"),      # Empty jug2
            ((a - min(a, y - b), b + min(a, y - b)), "Pour Jug 1 -> Jug 2"),  # Pour jug1 -> jug2
            ((a + min(b, x - a), b - min(b, x - a)), "Pour Jug 2 -> Jug 1")   # Pour jug2 -> jug1
        ]
        
        # Add valid next states to the queue
        for (state, action) in next_states:
            queue.append((state, path + [(a, b, action)]))
    
    return None  # No solution found
x = 3  # Capacity of jug1
y = 5  # Capacity of jug2
target = 4  # Target amount of water

# Check with BFS and DFS
print("Can measure using BFS:", bfs_water_jug(x, y, target,0))
print("Can measure using DFS:", bfs_water_jug(x, y, target,-1))
```

```python
#4th question Missionaries

from collections import deque

def is_valid(state):
    """Check if the state is valid (no missionaries outnumbered by cannibals)."""
    m1, c1, boat, m2, c2 = state
    if (m1 < c1 and m1 > 0) or (m2 < c2 and m2 > 0):
        return False
    return True

def get_next_states(state):
    """Generate all possible valid next states from the current state."""
    m1, c1, boat, m2, c2 = state
    moves = [(1, 0), (2, 0), (0, 1), (0, 2), (1, 1)]  # Possible moves
    next_states = []
    
    for m, c in moves:
        if boat == 1:  # Moving from left to right
            new_state = (m1 - m, c1 - c, 0, m2 + m, c2 + c)
        else:  # Moving from right to left
            new_state = (m1 + m, c1 + c, 1, m2 - m, c2 - c)
        
        if all(x >= 0 for x in new_state) and is_valid(new_state):
            next_states.append(new_state)
    
    return next_states

def search_missionaries_cannibals(strategy='BFS'):
    """Solve the problem using BFS or DFS."""
    start = (3, 3, 1, 0, 0)  # (Missionaries left, Cannibals left, Boat, Missionaries right, Cannibals right)
    goal = (0, 0, 0, 3, 3)
    
    queue = deque([(start, [])]) if strategy == 'BFS' else [(start, [])]
    visited = set()
    
    while queue:
        state, path = queue.popleft() if strategy == 'BFS' else queue.pop()
        
        if state in visited:
            continue
        visited.add(state)
        new_path = path + [state]
        
        if state == goal:
            return new_path  # Return the solution path
        
        for next_state in get_next_states(state):
            queue.append((next_state, new_path))
    
    return None  # No solution found

# Example Usage
print("BFS Solution:")
solution_bfs = search_missionaries_cannibals('BFS')
for step in solution_bfs:
    print(step)

print("\nDFS Solution:")
solution_dfs = search_missionaries_cannibals('DFS')
for step in solution_dfs:
    print(step)
```

```python
#5th question

ACTIONS = ['left', 'right', 'suck']

def transition(state, action):
    """Returns the next state after performing an action."""
    agent_pos, left_dirt, right_dirt = state
    if action == 'left':
        return (0, left_dirt, right_dirt) if agent_pos == 1 else state
    elif action == 'right':
        return (1, left_dirt, right_dirt) if agent_pos == 0 else state
    elif action == 'suck':
        if agent_pos == 0 and left_dirt == 1:
            return (0, 0, right_dirt)
        elif agent_pos == 1 and right_dirt == 1:
            return (1, left_dirt, 0)
    return state  # No change if action is invalid

def goal_state(state):
    """Checks if both left and right are clean."""
    return state[1] == 0 and state[2] == 0

def search(initial, strategy='BFS'):
    """Performs BFS or DFS based on the strategy."""
    queue = [(initial, 0, [], [])]  # (state, cost, path, actions)
    visited = set([initial])
    
    while queue:
        state, cost, path, actions = queue.pop(0) if strategy == 'BFS' else queue.pop(-1)
        
        if goal_state(state):
            return path + [state], actions, cost
        
        for action in ACTIONS:
            next_state = transition(state, action)
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, cost + 1, path + [state], actions + [action]))
    
    return -1, -1, -1  # No solution

# Example Usage
initial_state = (0, 1, 1)  # Agent on left, both sides dirty
print("\nBFS Exploration:")
bfs_path, bfs_actions, bfs_cost = search(initial_state, 'BFS')
if bfs_path != -1:
    print(f" BFS Path to goal state: {bfs_path}")
    print(f" BFS Actions taken: {bfs_actions}")
    print(f" BFS Path cost: {bfs_cost}")
else:
    print(" No path found")

print("\nDFS Exploration:")
dfs_path, dfs_actions, dfs_cost = search(initial_state, 'DFS')
if dfs_path != -1:
    print(f" DFS Path to goal state: {dfs_path}")
    print(f" DFS Actions taken: {dfs_actions}")
    print(f" DFS Path cost: {dfs_cost}")
else:
    print(" No path found")
```

```python
#6th question

def greed(cost, h, source, dest):
    op = []  # List to store the traversal path
    c = 0    # Total path cost

    while source != dest:
        op.append(source)
        children = []
        children_f = []

        # Find all children nodes and their heuristic values
        for i in range(len(cost[source])):
            if cost[source][i] > 0:
                children.append(i)
                children_f.append(h[i])

        # If no children, break (to prevent IndexError)
        if not children:
            break

        # Choose the child with the minimum heuristic value
        index = children_f.index(min(children_f))
        next_node = children[index]
        
        # Update cost
        c += cost[source][next_node]
        source = next_node

    op.append(source)
    return op, c

# Input Section
n = int(input("Enter number of nodes: "))
cost = [[0] * n for _ in range(n)]

print("Enter the names of nodes and the path cost as follows:")
print("<from_node> <to_node> <cost> (Enter 'x' to stop)")
while True:
    s = input("Enter: ")
    if s.lower() == 'x':
        break
    s = s.split()
    cost[int(s[0])-1][int(s[1])-1] = int(s[2])

h = list(map(int, input("Enter heuristic values: ").split()))
source = int(input("Enter source node: ")) - 1
dest = int(input("Enter destination node: ")) - 1

# Run Greedy Best-First Search
o, c = greed(cost, h, source, dest)

# Print Results
print("Nodes Traversed:")
print(" -> ".join(str(node + 1) for node in o))
print(f"Total Path Cost: {c}")




def Astar(cost,h,source,dest):
    op=[]
    c=[0]
    while source!=dest:
        op.append(source)
        if len(op)>1:
            c.append(cost[op[-2]][op[-1]])
        children=[]
        children_f=[]
        for i in range(len(cost[0])):
            if cost[source][i]>0:
                children.append(i)
                children_f.append(sum(c)+h[i]+cost[source][i])
        if len(children)>0:
            index = children_f.index(min(children_f))
        source=children[index]

    op.append(source)
    c.append(cost[op[-2]][op[-1]])
    return op,c

n=int(input("enter no of nodes:"))
cost=[[0]* n for i in range(n)]
for i in range(n):
    print(i+1,end=" ")
print("enter the names of nodes and the path cost as follows:")
print("<from_node><to_node><cost>")
while True:
    s=input("enter")
    if s=='x':
        break
    s=s.split()
    cost[int(s[0])-1][int(s[1])-1]=int(s[2])
h=list(map(int,input("enter heuristic values:").split()))
source=int(input("enter source node:"))-1
dest=int(input("enter destination node:"))-1
op,c=Astar(cost,h,source,dest)
print("Nodes Traversed:")
for i in op[:-1]:
    print(i+1,end='->')
print(op[-1] + 1)
print("cost of traverse is:",sum(c))
```

```python
#8th question

from collections import deque

def get_neighbors(state):
    moves = []
    index = state.index("_")

    swaps = {
        "L": index - 1 if index % 3 != 0 else None,
        "R": index + 1 if index % 3 != 2 else None,
        "U": index - 3 if index >= 3 else None,
        "D": index + 3 if index <= 5 else None,
    }
    
    for move, new_index in swaps.items():
        if new_index is not None:
            new_state = list(state)
            new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
            moves.append("".join(new_state))
    
    return moves

def bfs_solver(start, goal):
    queue = deque([(start, [])])  # (current state, path to reach it)
    visited = set()

    while queue:
        state, path = queue.popleft()
        if state == goal:
            return path + [state]  # Return the sequence of moves
        if state in visited:
            continue
        visited.add(state)
        for neighbor in get_neighbors(state):
            queue.append((neighbor, path + [state]))
    return None  # No solution found
```

```python
#9th question

def cost(H, condition, weight=1):
    costs = {}
    if 'AND' in condition:
        AND_nodes = condition['AND']
        costs[' AND '.join(AND_nodes)] = sum(H[node] + weight for node in AND_nodes)
    if 'OR' in condition:
        OR_nodes = condition['OR']
        costs[' OR '.join(OR_nodes)] = min(H[node] + weight for node in OR_nodes)
    return costs

def update_cost(H, Conditions, weight=1):
    for node in reversed(Conditions.keys()):
        H[node] = min(cost(H, Conditions[node], weight).values())
    return {node: cost(H, Conditions[node], weight) for node in Conditions}

def shortest_path(start, Updated_cost):
    if start not in Updated_cost:
        return start
    min_cost_key = min(Updated_cost[start], key=Updated_cost[start].get)
    next_nodes = min_cost_key.split(' AND ')
    if len(next_nodes) == 1:
        return start + " <-- " + shortest_path(next_nodes[0], Updated_cost)
    else:
        return f"{start} <-- ({min_cost_key}) [{shortest_path(next_nodes[0], Updated_cost)} + {shortest_path(next_nodes[1], Updated_cost)}]"

# Input
H = eval(input('Enter heuristic values as a dictionary: '))
Conditions = eval(input('Enter conditions as a dictionary: '))
Updated_cost = update_cost(H, Conditions)
print('Updated Costs:', Updated_cost)
print('Shortest Path:', shortest_path('A', Updated_cost))
```

```python
#11th question


import itertools

def number(word,digit_map):
    return int(''.join(str(digit_map[letter]) for letter in word))

def solve_cryptarithmetic(puzzle):
    words=puzzle.split()
    unique_characters=set(''.join(words))
    if len(unique_characters)>10:
        return '\n Valid Puzzle: more than 10 unique characters'
    leading_characters = set(word[0] for word in words)
    if len(leading_characters)>2:
        return 'Invalid Puzzle: more than 2 words start with the same character'
    for digits in itertools.permutations(range(10),len(unique_characters)):
        digit_map=dict(zip(unique_characters,digits))
        if all(digit_map[word[0]]!=0 for word in leading_characters):
            if sum(number(word,digit_map) for word in words[:-1])== number(words[-1],digit_map):
                return digit_map
    return "No Solution Found"

puzzle = input("enter the cruptarithmetic puzzle:")
solution = solve_cryptarithmetic(puzzle)
print("solution:",solution)
```
